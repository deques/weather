```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('weather_data.csv')
df = df.iloc[:, 1:]
df["date"] = pd.to_datetime(df["date"])
df
```

```python
# Plot the data
plt.figure(figsize=(10, 5))  # Set figure size
plt.plot(df["date"], df["temperature_2m_max"], label="Temperatur (Â°C)", color="blue", marker="o")

# Format the X-axis
plt.xlabel("Datum")
plt.ylabel("Temperatur (Â°C)")
plt.title("Temperatur i NorrkÃ¶ping under 2024")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend()
plt.grid()

plt.show()
```

```python
# Create figure and axis
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot Temperature on Primary Y-Axis
ax1.plot(df["date"], df["temperature_2m_max"], label="Max Temperatur (Â°C)", color="red", marker="o")
ax1.plot(df["date"], df["temperature_2m_min"], label="Min Temperatur (Â°C)", color="blue", marker="o")
ax1.set_xlabel("Datum")
ax1.set_ylabel("Temperatur (Â°C)")
ax1.tick_params(axis="y", labelcolor="red")

# Create a second Y-axis for precipitation
ax2 = ax1.twinx()
ax2.bar(df["date"], df["precipitation_sum"], label="NederbÃ¶rd (mm)", color="black", alpha=0.5)
ax2.set_ylabel("NederbÃ¶rd (mm)")
ax2.tick_params(axis="y", labelcolor="blue")

# Formatting
plt.title("Temperatur & nederbÃ¶rd i NorrkÃ¶ping")
fig.autofmt_xdate()  # Rotate x-axis labels
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.grid()
```
